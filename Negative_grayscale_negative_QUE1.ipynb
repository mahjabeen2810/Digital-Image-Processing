```python
from PIL import Image,ImageOps
def display_image(image_path):
    img=Image.open(image_path)
    display(img)
    
    negative_image=ImageOps.invert(img)
    display(negative_image)
    
    gray_scale=ImageOps.grayscale(img)
    display(gray_scale)
    
    gray_scale_negative=ImageOps.invert(gray_scale)
    display(gray_scale_negative)
image_path="sunflower.jpg"
display_image(image_path)
```
