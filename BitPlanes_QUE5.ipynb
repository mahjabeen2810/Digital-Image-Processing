```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread("lotus.jpg",0)
bit_planes=[]

for i in range (8):
    planes=np.zeros_like(img,dtype=np.uint8)
    planes[img & (1<<i)!=0]=255
    bit_planes.append(planes)
    
plt.figure(figsize=(12,6))

for i in range (8):
    plt.subplot(2,4,i+1)
    plt.imshow(bit_planes[i],cmap="gray")
    plt.title(f"BIT_PLANE : {i+1}")
plt.tight_layout()
plt.show()
```
